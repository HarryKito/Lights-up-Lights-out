```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
on_light = cv2.imread('resources/L_on.jpg')
on_light = cv2.cvtColor(on_light,cv2.COLOR_BGR2HSV_FULL)
off_light = cv2.imread('resources/L_off.jpg')
off_light = cv2.cvtColor(off_light,cv2.COLOR_BGR2HSV_FULL)
```

```python
bright_arr = np.asarray(on_light)
darker_arr = np.asarray(off_light)
```

```python
print(bright_arr)
```

```python
# light on
plt.subplot(2,2,1)
plt.title("ON")
plt.imshow(on_light)
plt.xticks([]), plt.yticks([])
# light off
plt.subplot(2,2,2)
plt.title("OFF")
plt.imshow(off_light)
plt.xticks([]), plt.yticks([])
# light on (graph)
plt.subplot(2,2,3)
plt.imshow(off_light)
plt.xticks([]), plt.yticks([])
# light off (graph)
plt.subplot(2,2,4)
plt.imshow(off_light)
plt.xticks([]), plt.yticks([])
```
