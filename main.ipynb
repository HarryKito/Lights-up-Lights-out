```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
on_light = cv2.imread('L_on.jpg')
on_light = cv2.cvtColor(on_light,cv2.COLOR_BGR2RGB)
off_light = cv2.imread('L_off.jpg',cv2.COLOR_BGR2RGB)
```

```python
# light on
plt.subplot(2,2,1)
plt.title("ON")
plt.imshow(on_light)
plt.xticks([]), plt.yticks([])
# light off
plt.subplot(2,2,2)
plt.title("OFF")
plt.imshow(off_light)
plt.xticks([]), plt.yticks([])
# light on (graph)
plt.subplot(2,2,3)
plt.imshow(off_light)
plt.xticks([]), plt.yticks([])
# light off (graph)
plt.subplot(2,2,4)
plt.imshow(off_light)
plt.xticks([]), plt.yticks([])
```
